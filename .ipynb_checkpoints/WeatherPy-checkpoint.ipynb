### Analysis
* As we approach the equator, the weather becomes warmer.
* In terms of cloudiness and latitude, there does not seem to be a strong correlation.
* In terms of wind speed and latitude, there does not seem to be a strong correlation.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from pprint import pprint 
import datetime

# Import API key
from config import api_keys

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
#api.openweathermap.org/data/2.5/weather?q={city name}&appid={API key}
# Save config information
url = "http://api.openweathermap.org/data/2.5/weather?"
parameters = f"appid={api_keys}"

#declare lists
city_name = []
temp = []
hum = []
cloudy = []
wind = []
lat = []
date = []

# Build query URL
for x in cities:
    query_url = url + "q=" + x + "&" + parameters
    response = requests.get(query_url).json()
    try:
        city_name.append(response["name"])
        temp.append(response["main"]["temp"])
        hum.append(response["main"]["humidity"])
        cloudy.append(response["clouds"]["all"])
        wind.append(response["wind"]["speed"])
        lat.append(response["coord"]["lat"])
        date.append(response["dt"])
    except:
        pass
    
pprint(response)
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#build dataframe
city_data_df = pd.DataFrame({
    "City":city_name,
    "Latitude":lat,
    "Temperature":temp,
    "Humidity":hum,
    "Cloudiness":cloudy,
    "Wind Speed":wind,
    "Date":date
})

#convert UNIX date
city_data_df['Date'] = pd.to_datetime(city_data_df['Date'],unit='s')
date_final = city_data_df['Date'][0]


#write to output file
city_data_df.to_csv(output_data_file)
date_final = str(date_final)
date_final = date_final[0:10]

#display dataframe
city_data_df
```
